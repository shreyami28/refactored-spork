<h2> Data Cleanup </h2>

Import numpy, pandas, matplotlib, and seaborn

(Feel free to import other libraries!)

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

Load in the "states_edu.csv" dataset and take a look at the head of the data

```python
df= pd.read_csv('../data/states_edu.csv')
df.head()
```

You should always familiarize yourself with what each column in the dataframe represents. \ Read about the states_edu dataset here: https://www.kaggle.com/noriuk/us-education-datasets-unification-project

Use this space to rename columns, deal with missing data, etc. _(optional)_

```python
# Renaming Columns
df.rename ({
   'GRADES_PK_G':'ENROLL_PREK',
    'GRADES_KG_G':'ENROLL_KINDER',
    'GRADES_4_G':'ENROLL_4',
    'GRADES_8_G':'ENROLL_8',
    'GRADES_12_G':'ENROLL_12',
    'GRADES_1_8_G':'ENROLL_PRIMARY',
    'GRADES_9_12_G':'ENROLL_HS',
    'GRADES_ALL_G':'ENROLL_ALL',
    'ENROLL':'ENROLL_ALL_EST'
    },
    axis=1, inplace=True)
```

```python
df.head()
```

```python
# Choosing to look at Math score for 8th graders 
# Drop rows with at least 1 nan in "AVG_MATH_8_SCORE" column  
data_without_NaN= df.dropna(subset=["AVG_MATH_8_SCORE"])
data_without_NaN.head()
```

```python
# Removing columns including grades not 8 and not Math 
data_8th_grade_Math= data_without_NaN.drop(columns=['AVG_MATH_4_SCORE', 'AVG_READING_4_SCORE', 'AVG_READING_8_SCORE'])
data_8th_grade_Math.head()
```

How many different years of data are in our dataset? Use a pandas function.

```python
# Set year to a variable, then used unique to find unique year values
column_values = data_8th_grade_Math[["YEAR"]].values.ravel()
unique_year_values= pd.unique(column_values)
print(unique_year_values)
```

Let's compare Michigan to Ohio. Which state has the higher average predictor score across all years?

```python
# Found which average math 8 score is higher between Michigan to Ohio
Michigan_Ohio_comparison_score= data_8th_grade_Math[data_8th_grade_Math.STATE.isin(["MICHIGAN",'OHIO'])].groupby('STATE').AVG_MATH_8_SCORE.mean() 
print(Michigan_Ohio_comparison_score)
# Ohio has the higher average score 
```

Find the average for your pedictor score across all states in 2019

```python
Average_Predictor= data_8th_grade_Math[data_8th_grade_Math.YEAR.isin([2019])].groupby('YEAR').AVG_MATH_8_SCORE.mean()
print(Average_Predictor)
Predictor= data_8th_grade_Math.groupby("YEAR")
Predictor.AVG_MATH_8_SCORE.mean()
# In 2019 the average Math score was 281.26 
```

Find the maximum predictor score for every state. Hint: there's a function that allows you to do this easily

<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
X = data_8th_grade_Math[['SUPPORT_SERVICES_EXPENDITURE','ENROLL_8','LOCAL_REVENUE']].dropna()
y = data_8th_grade_Math.loc[X.index]['AVG_MATH_8_SCORE']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=.3, random_state=0)
```

<h2> Prediction </h2>

ML Models Resource: https://medium.com/@vijaya.beeravalli/comparison-of-machine-learning-classification-models-for-credit-card-default-data-c3cf805c9a5a

Chosen ML task: **<REGRESSION/CLASSIFICATION>**

```python
# import your sklearn class here
from sklearn.linear_model import LinearRegression
```

```python
# create your model here
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
# FOR REGRESSION ONLY: (pick a single column to visualize results)

# Results from this graph _should not_ be used as a part of your results -- it is just here to help with intuition. 
# Instead, look at the error values and individual intercepts.


col_name = "LOCAL_REVENUE"
col_index = X_train.columns.get_loc(col_name)

f = plt.figure(figsize=(12,6))
plt.scatter(X_train[col_name], y_train, color = "red")
plt.scatter(X_train[col_name], model.predict(X_train), color = "green")
plt.scatter(X_test[col_name], model.predict(X_test), color = "blue")

new_x = np.linspace(X_train[col_name].min(),X_train[col_name].max(),200)
intercept = model.predict([X_train.sort_values(col_name).iloc[0]]) - X_train[col_name].min()*model.coef_[col_index]
plt.plot(new_x, intercept+new_x*model.coef_[col_index])

plt.legend(['controlled model','true training','predicted training','predicted testing'])
plt.xlabel(col_name)
plt.ylabel("8th Grade Math Scores")
```

```python
print(model.intercept_)
print(model.coef_)
```

```python
model.score(X_test, y_test)
```
